```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('Dataset/titanic.csv')
```

```python
df.head()
```

```python
plt.hist(df['Sex'])
plt.show()
```

```python
df['Embarked'].fillna(df['Embarked'].mode().iloc[0], inplace=True)
```

```python
df['Embarked'].unique()
```

```python
df['Embarked']
```

```python
plt.hist(df['Embarked'])
plt.show()
```

```python
plt.pie([df['Sex'].value_counts()['male'], df['Sex'].value_counts()['female']], labels=['male', 'female'], autopct='%.2f%%')
plt.show()
```

```python
plt.hist(df['Survived'])
plt.show()
```

```python
plt.pie([df['Survived'].value_counts()[0], df['Survived'].value_counts()[1]], labels=['dead', 'alive'], autopct='%.2f%%')
plt.show()
```
