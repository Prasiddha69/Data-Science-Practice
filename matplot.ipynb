```python
companies = ['google', 'amazon', 'fb']
revenue = [100, 58, 34]
```

```python
plt.bar(companies, revenue)
plt.show()
```

```python
plt.bar(companies, revenue, color='r')
plt.show()
```

```python
plt.bar(companies, revenue, color=['r', 'b', 'g'])
plt.xlabel('company names')
plt.ylabel('revenue')
plt.title('companies vs revenue')
plt.show()
```

```python
plt.pie(revenue, colors=['r', 'g', 'b'], labels=companies, autopct='%.2f%%')
plt.show()
```
