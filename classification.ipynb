```python
import pandas as pd
 # logistic regression
```

```python
df= pd.read_csv("titanic.csv")
```

```python
df.head(4)
```

```python
df=df[["Sex","Age","Survived"]]
```

```python
df
```

```python
df.isnull().sum()
```

```python
mean_age=round(df.Age.mean(),1)
```

```python
df.Age.fillna(mean_age,inplace=True)
```

```python
df
```

```python
gender=pd.get_dummies(df.Sex,drop_first=True,dtype=int)
```

```python
gender
```

```python
df
```

```python
df=pd.concat([df,gender],axis=1)
```

```python
df
```

```python
df.drop("Sex",axis=1,inplace=True)
```

```python
df
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[["Age","male"]],df.Survived, test_size=0.20, random_state=42)
```

```python
X_test
```

```python
model.fit(X_train,y_train)
```

```python
predict=model.predict(X_test)
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,predict)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
pd.DataFrame(confusion_matrix(y_test,predict))
```
